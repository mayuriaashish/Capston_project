```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
#declaring the list of empty variables, So that we can append the data overall
SNO = []
Movie_Name = []
Director_Name = []
Duration = []
Year = []
Rating = []
Metascore = []
```

```python
#creating an array of values and passing it in the url for dynamic webpages
pages = np.arange(1,1771,70)
```

```python
#the whole core of the script
for page in pages:
    url = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start=" + str(page) + "&ref_=adv_nxt"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    movie_data = soup.findAll('div', class_='lister-item mode-advanced')

    for store in movie_data:
        sno_element = store.find("span", class_="lister-item-index unbold text-primary")
        sno = sno_element.text.strip(".") if sno_element else ""
        SNO.append(sno)

        name_element = store.h3.a
        name = name_element.text if name_element else ""
        Movie_Name.append(name)

        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
        Director_Name.append(cast[0])

        duration_element = store.find("span", class_="runtime")
        duration = duration_element.text if duration_element else ""
        Duration.append(duration)

        year_element = store.find("span", class_="lister-item-year")
        year = year_element.text.strip("()") if year_element else ""
        Year.append(year)

        rating_element = store.find("div", class_="ratings-imdb-rating").strong
        rating = rating_element.text if rating_element else ""
        Rating.append(rating)

        metascore_element = store.find("span", class_="metascore")
        metascore = metascore_element.text.strip() if metascore_element else ""
        Metascore.append(metascore)

    sleep(randint(2, 8))

data = {
    'SNO': SNO,
    'Movie Name': Movie_Name,
    'Director Name': Director_Name,
    'Duration': Duration,
    'Year': Year,
    'Rating': Rating,
    'Metascore': Metascore
}

```

```python
# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('movie_details1.csv', index=False)

print("Data saved to movie_details1.csv")
```
