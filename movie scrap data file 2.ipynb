```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
#declaring the list of empty variables, So that we can append the data overall
Movie_Name = []
Stars_Name = []
Votes = []
Genre = []
Gross_Collection = []
Popularity = []
Certification = []
```

```python
#creating an array of values and passing it in the url for dynamic webpages
pages = np.arange(1,1771,70)
```

```python
#the whole core of the script
for page in pages:
    url = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start=" + str(page) + "&ref_=adv_nxt"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    movie_data = soup.findAll('div', class_='lister-item mode-advanced')

    for store in movie_data:
        name_element = store.h3.a
        name = name_element.text if name_element else ""
        Movie_Name.append(name)
        
        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
        Stars_Name.append([x.strip() for x in cast[1].split(",")])

        genre_element = store.find("span", class_= "genre").text.replace('\n', '')
        Genre.append(genre_element)
        
        value = store.find_all('span', attrs = {'name': 'nv'})
    
        vote = value[0].text
        Votes.append(vote)
    
        grosses = value[1].text if len(value) >1 else '*****'
        Gross_Collection.append(grosses)

        popularity_element = store.find("div", class_="ratings-imdb-rating").strong
        popularity = popularity_element.text if popularity_element else ""
        Popularity.append(popularity)

        certification_element = store.find("span", class_="certificate")
        certification = certification_element.text.strip() if certification_element else ""
        Certification.append(certification)

    sleep(randint(2, 8))

data = {
    'Movie Name': Movie_Name,
    'Stars Name': Stars_Name,
    'Votes': Votes,
    'Genre': Genre,
    'Gross Collection': Gross_Collection,
    'Popularity': Popularity,
    'Certification': Certification
}
```

```python
data = {
    'Movie Name': Movie_Name,
    'Stars Name': Stars_Name,
    'Votes': Votes,
    'Genre': Genre,
    'Gross Collection': Gross_Collection,
    'Popularity': Popularity,
    'Certification': Certification
}
```

```python
# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('movie_details3.csv', index=False)

print("Data saved to movie_details3.csv")
```
